# Newton's law of universal gravitation
F = (Gm1m2) / (r^2)

```python
import matplotlib.pyplot

def drawGraph(x,y):
    matplotlib.pyplot.plot(x,y,'o')
    matplotlib.pyplot.xlabel("Distance [m]")
    matplotlib.pyplot.ylabel("Gravity [N]")
    matplotlib.pyplot.title("Dependence of gravity on distance")
    matplotlib.pyplot.show()

def generateFr():
    r = range(100,1001,50)
    F = []
    G = 6.674*(10**-11)
    m1 = 0.5 #object1 weight
    m2 = 1.5 #object2 weight
    for dist in r:
        force = G*(m1*m2) / (dist**2)
        F.append(force)
    drawGraph(r,F)

generateFr()
```
