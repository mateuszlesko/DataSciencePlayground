```python
import matplotlib.pyplot
```

The basic linear plot that visualise average temperatures each year in NYC

```python
nyc_temps = [12.17,13.5,13.56,11.89,12.5,13.22,13.78,12.78,12.94,12.22,13.72,13.56,14.06]
years = range(2000,2013)
matplotlib.pyplot.plot(years,nyc_temps,marker="o")

```

Plot that visualise average mouthly temperatures in NYC from 2000, 2006 and 2012 as a seperate series

```python
nyc_2000 = [-0.39, 2.94, 8.44, 10.56,17.5,21.83,22.39,22.61,19.98,13.89,7.39,-0.5]
nyc_2006 = [4.94, 2.06, 6.44, 13.56,17.28,21.67,25.5,24.61,19.22,13.44,11.39,6.5]
nyc_2012 = [2.94, 4.06, 10.44, 12.56,18.28,21.67,26.5,24.61,20.22,14.44,6.39,5.5]
months = range(1,13)
matplotlib.pyplot.subplot(312)
matplotlib.pyplot.legend([200,2006,2012])
matplotlib.pyplot.plot(months,nyc_2000)
matplotlib.pyplot.plot(months,nyc_2006)
matplotlib.pyplot.plot(months,nyc_2012)

```
